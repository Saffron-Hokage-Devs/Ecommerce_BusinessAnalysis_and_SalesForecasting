```python
import numpy as np, pandas as pd

#ignore warnings ...
import warnings
warnings.filterwarnings('ignore')
```

```python

# Read the CSV file
df = pd.read_csv('archive.csv')
df.head(2)
```

```python
df.shape
```

```python
df = df.iloc[:, :-5]
df = df.dropna(how = 'all')
print(df.shape)
df.head(3)
```

```python
# Convert the 'dates' column to datetime format
df['created_at'] = pd.to_datetime(df['created_at'], dayfirst=True, errors='coerce')
df['Working Date'] = pd.to_datetime(df['Working Date'], dayfirst=True, errors='coerce')

# Convert the 'dates' column to the desired format
df['created_at'] = df['created_at'].dt.strftime('%d-%m-%Y')
df['Working Date'] = df['Working Date'].dt.strftime('%d-%m-%Y')

df.head()
```

```python
df.info()
```

```python
# replacing NaN with  "complete" in "status"
df["status"].fillna("complete", inplace = True)
print("null values in status attribute = ", df["status"].isnull().sum())

#replacing NaN with "Mobiles & Tablets" in "category_name_1"
df["category_name_1"].fillna("Mobiles & Tablets", inplace = True)
print("null values in category_name_1 attribute = ", df["category_name_1"].isnull().sum())
```

```python
# Looking out for other missing values...
null_columns = df.columns[df.isnull().any()]
null_counts = df[null_columns].isnull().sum()

for column in null_columns:
    print(f"Column: {column}, Null Count: {null_counts[column]}")
```

```python
df["sales_commission_code"].fillna(r"\N", inplace=True)
df= df.dropna()
df.shape
```

```python

# Convert the date columns to datetime type if they are not already
df['created_at'] = pd.to_datetime(df['created_at'])
df['Working Date'] = pd.to_datetime(df['Working Date'])

# Create a new column 'differing_values' to store the comparison result
df['differing_values'] = df['created_at'] != df['Working Date']

# Filter the dataframe to include only rows with differing values
differing_rows = df[df['differing_values']]

# Print the resulting rows
print(differing_rows)
```
